```python
import pygrib
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
import numpy as np
import os
%matplotlib inline
```

```python
grbs = pygrib.open(os.getcwd() + '/DATA/filtered_weather_data/namanl/namanl_218_20130101_1200_003.grb')
uin = grbs.select(shortName='u', typeOfLevel='isobaricInhPa', level = lambda l: l >= 150 and l <= 1000)
grbs.close()
```

```python
Lat = uin[0].data()[1]
Lon = uin[0].data()[2]
Grid = np.dstack((Lon.flatten(),Lat.flatten()))[0]
```

```python
plt.figure(figsize=(12,10))
# Lambert Conformal Conic map.
m = Basemap(width=12000000,height=9000000,
            rsphere=(6378137.00,6356752.3142),\
            resolution='l',area_thresh=1000.,projection='lcc',\
            lat_1=40.,lat_2=60.,lat_0=45,lon_0=-100.)
m.drawcoastlines()
m.drawcountries()
m.drawmapboundary()
m.fillcontinents(color='#d3d3d3')
m.drawparallels(np.arange(10,60,20),labels=[1,1,0,0])
m.drawmeridians(np.arange(-180,0,30),labels=[0,0,0,1])

x,y = m(Grid[:,0], Grid[:,1])
plt.scatter(x,y, s = 0.01, zorder = 10)
plt.show()
```
